# ***Bresenham's Circle Drawing Algorithm***

```python
import matplotlib.pyplot as plt

def draw_circle_bresenham(xc, yc, r):

    x = 0
    y = r
    d = 3 - 2 * r

    x_points = []
    y_points = []

    def plot_circle_points(xc, yc, x, y):

        x_points.extend([xc + x, xc - x, xc + x, xc - x, xc + y, xc - y, xc + y, xc - y])
        y_points.extend([yc + y, yc + y, yc - y, yc - y, yc + x, yc + x, yc - x, yc - x])

    plot_circle_points(xc, yc, x, y)

    while x < y:
        x += 1
        if d < 0:
            d = d + 4 * x + 6
        else:
            y -= 1
            d = d + 4 * (x - y) + 10
        plot_circle_points(xc, yc, x, y)

    plt.figure(figsize=(8, 8))
    plt.plot(x_points, y_points, 'ro', markersize=2)
    plt.title(f"Bresenham's Circle: Center=({xc},{yc}), Radius={r}")
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')
    plt.grid(True)
    plt.axis('equal')
    plt.show()

def main():
    center_x, center_y = 0, 0
    radius = 1000
    draw_circle_bresenham(center_x, center_y, radius)

if __name__ == "__main__":
    main()
```

# ***Bresenham's Line Drawing Algorithm***

```python
import matplotlib.pyplot as plt

def draw_line_Bresenham(x1, y1, x2, y2):
    points = []
    dx = x2 - x1
    dy = y2 - y1

    x, y = x1, y1

    points.append((x, y))

    p = 2 * dy - dx

    for _ in range(dx):
        x += 1
        if p < 0:
            p = p + 2 * dy
        else:
            p = p + 2 * dy - 2 * dx
            y += 1
        points.append((x, y))

    return points

def plot_line(points, title):
    x_values, y_values = zip(*points)
    plt.plot(x_values, y_values, marker='o')
    plt.title(title)
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')
    plt.grid(True)
    plt.show()

# Example usage:
x1, y1 = 2, 3
x2, y2 = 9, 8

# Bresenham's algorithm
bresenham_points = draw_line_Bresenham(x1, y1, x2, y2)
plot_line(bresenham_points, 'Bresenham\'s Line Drawing')
```

# ***Digital Differential Analyzer Line Drawing Algorithm***

```python
import matplotlib.pyplot as plt
def draw_line_DDA(x1, y1, x2, y2):
    points = []
    dx = x2 - x1
    dy = y2 - y1
    steps = max(abs(dx), abs(dy))

    x_increment = dx / steps
    y_increment = dy / steps

    x, y = x1, y1

    for _ in range(steps):
        points.append((round(x), round(y)))
        x += x_increment
        y += y_increment

    return points

def plot_line(points, title):
    x_values, y_values = zip(*points)
    plt.plot(x_values, y_values, marker='o')
    plt.title(title)
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')
    plt.grid(True)
    plt.show()

# Example usage:
x1, y1 = 2, 3
x2, y2 = 9, 8

# DDA algorithm
dda_points = draw_line_DDA(x1, y1, x2, y2)
plot_line(dda_points, 'DDA Line Drawing')
```

# ***Midpoint Circle Drawing Algorithm***

```python
import matplotlib.pyplot as plt

def draw_circle_midpoint(xc, yc, r):
    """
    Implements Midpoint Circle Drawing Algorithm
    xc, yc: center coordinates
    r: radius
    """
    points = []
    x = 0
    y = r
    p = 1 - r  # Initial decision parameter

    # Store the first set of points
    add_circle_points(xc, yc, x, y, points)

    while x < y:
        x += 1
        if p < 0:
            p += 2 * x + 1
        else:
            y -= 1
            p += 2 * (x - y) + 1

        add_circle_points(xc, yc, x, y, points)

    return points

def add_circle_points(xc, yc, x, y, points):
    """
    Add points for all eight octants of the circle
    """
    points.extend([
        (xc + x, yc + y), (xc - x, yc + y),
        (xc + x, yc - y), (xc - x, yc - y),
        (xc + y, yc + x), (xc - y, yc + x),
        (xc + y, yc - x), (xc - y, yc - x)
    ])

def plot_circle(points, title):
    """
    Plot the circle points
    """
    x_values, y_values = zip(*points)
    plt.plot(x_values, y_values, 'bo')  # 'bo' for blue dots
    plt.title(title)
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')
    plt.grid(True)
    plt.axis('equal')  # Make the plot circular
    plt.show()

def main():
    # Example usage
    center_x, center_y = 0, 0  # Circle center
    radius = 5  # Circle radius

    # Midpoint Circle algorithm
    circle_points = draw_circle_midpoint(center_x, center_y, radius)
    plot_circle(circle_points, 'Midpoint Circle Drawing Algorithm')

if __name__ == "__main__":
    main()
```

# ***Midpoint Line Drawing Algorithm***

```python
import matplotlib.pyplot as plt

def draw_line_midpoint(x1, y1, x2, y2):
    points = []
    dx = x2 - x1
    dy = y2 - y1

    x, y = x1, y1
    points.append((x, y))

    d = dy - (dx / 2)
    while x < x2:
        x += 1
        if d < 0:
            d = d + dy
        else:
            y += 1
            d = d + (dy - dx)
        points.append((x, y))
    return points

def plot_line(points, title):
    x_values, y_values = zip(*points)
    plt.plot(x_values, y_values, marker='o')
    plt.title(title)
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')
    plt.grid(True)
    plt.show()

# Example usage:
x1, y1 = 3, 5
x2, y2 = 6, 7

midpoint_points = draw_line_midpoint(x1, y1, x2, y2)
plot_line(midpoint_points, "Midpoint Line Drawing")
```

# ***2D Translation***

```python
import numpy as np
import matplotlib.pyplot as plt

def draw_triangle(points, color='b', label=None):
    # Draw the triangle by connecting the points
    plt.plot([points[0][0], points[1][0]], [points[0][1], points[1][1]], color=color)
    plt.plot([points[1][0], points[2][0]], [points[1][1], points[2][1]], color=color)
    plt.plot([points[2][0], points[0][0]], [points[2][1], points[0][1]], color=color, label=label)

def translate_points(points, tx, ty):
    # Create translation matrix
    translation_matrix = np.array([[1, 0, tx],
                                 [0, 1, ty],
                                 [0, 0, 1]])

    # Convert points to homogeneous coordinates
    homogeneous_points = np.hstack((points, np.ones((3, 1))))

    # Apply translation
    translated_points = np.dot(homogeneous_points, translation_matrix.T)

    # Convert back to 2D coordinates
    return translated_points[:, :2]

# Define the original triangle vertices
original_triangle = np.array([
    [1, 1],    # Point 1
    [2, 3],    # Point 2
    [3, 1]     # Point 3
])

# Define translation values
tx, ty = 2, 2  # Translate 2 units in x and y directions

# Translate the triangle
translated_triangle = translate_points(original_triangle, tx, ty)

# Set up the plot
plt.figure(figsize=(10, 8))
plt.grid(True)

# Draw original triangle
draw_triangle(original_triangle, 'b', 'Original Triangle')

# Draw translated triangle
draw_triangle(translated_triangle, 'r', 'Translated Triangle')

# Add plot details
plt.title('2D Translation of a Triangle')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.axis('equal')  # Make the plot axes equal scale

# Show the plot
plt.show()
```

# ***2D Reflection***

```python
import numpy as np
import matplotlib.pyplot as plt

def draw_triangle(points, color='b', label=None):
    # Draw the triangle by connecting the points
    plt.plot([points[0][0], points[1][0]], [points[0][1], points[1][1]], color=color)
    plt.plot([points[1][0], points[2][0]], [points[1][1], points[2][1]], color=color)
    plt.plot([points[2][0], points[0][0]], [points[2][1], points[0][1]], color=color, label=label)

def reflect_points(points, axis='x'):
    # Create reflection matrix
    if axis.lower() == 'x':
        reflection_matrix = np.array([[1, 0, 0],
                                    [0, -1, 0],
                                    [0, 0, 1]])  # Reflection about X-axis
    else:
        reflection_matrix = np.array([[-1, 0, 0],
                                    [0, 1, 0],
                                    [0, 0, 1]])  # Reflection about Y-axis

    # Convert points to homogeneous coordinates
    homogeneous_points = np.hstack((points, np.ones((3, 1))))

    # Apply reflection
    reflected_points = np.dot(homogeneous_points, reflection_matrix.T)

    # Convert back to 2D coordinates
    return reflected_points[:, :2]

# Define the original triangle vertices
original_triangle = np.array([
    [1, 1],    # Point 1
    [2, 3],    # Point 2
    [3, 1]     # Point 3
])

# Reflect the triangle about X and Y axes
reflected_triangle_x = reflect_points(original_triangle, 'x')
reflected_triangle_y = reflect_points(original_triangle, 'y')

# Set up the plot
plt.figure(figsize=(12, 10))
plt.grid(True)

# Draw original and reflected triangles
draw_triangle(original_triangle, 'b', 'Original Triangle')
draw_triangle(reflected_triangle_x, 'r', 'Reflected about X-axis')
draw_triangle(reflected_triangle_y, 'g', 'Reflected about Y-axis')

# Add axes lines for reference
plt.axhline(y=0, color='k', linestyle='-', alpha=0.3)
plt.axvline(x=0, color='k', linestyle='-', alpha=0.3)

# Add plot details
plt.title('2D Reflection of a Triangle')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.axis('equal')  # Make the plot axes equal scale

# Set plot limits to show all triangles clearly


# Show the plot
plt.show()
```

# ***2D Shearing***

```python
import numpy as np
import matplotlib.pyplot as plt

def draw_triangle(points, color='b', label=None):
    # Draw the triangle by connecting the points
    plt.plot([points[0][0], points[1][0]], [points[0][1], points[1][1]], color=color)
    plt.plot([points[1][0], points[2][0]], [points[1][1], points[2][1]], color=color)
    plt.plot([points[2][0], points[0][0]], [points[2][1], points[0][1]], color=color, label=label)

def shear_points(points, shx=0, shy=0):
    # Create shearing matrix
    # shx: shearing factor in x-direction
    # shy: shearing factor in y-direction
    shear_matrix = np.array([[1, shx, 0],
                            [shy, 1, 0],
                            [0, 0, 1]])

    # Convert points to homogeneous coordinates
    homogeneous_points = np.hstack((points, np.ones((3, 1))))

    # Apply shearing
    sheared_points = np.dot(homogeneous_points, shear_matrix.T)

    # Convert back to 2D coordinates
    return sheared_points[:, :2]

# Define the original triangle vertices
original_triangle = np.array([
    [1, 1],    # Point 1
    [2, 3],    # Point 2
    [3, 1]     # Point 3
])

# Define shearing factors
shx = 0.5  # Shearing factor in x-direction
shy = 0.3  # Shearing factor in y-direction

# Apply shearing transformations
sheared_triangle_x = shear_points(original_triangle, shx=shx, shy=0)  # X-direction shearing
sheared_triangle_y = shear_points(original_triangle, shx=0, shy=shy)  # Y-direction shearing

# Set up the plot
plt.figure(figsize=(12, 10))
plt.grid(True)

# Draw original and sheared triangles
draw_triangle(original_triangle, 'b', f'Original Triangle')
draw_triangle(sheared_triangle_x, 'r', f'Sheared in X-direction (shx={shx})')
draw_triangle(sheared_triangle_y, 'g', f'Sheared in Y-direction (shy={shy})')

# Add axes lines for reference
plt.axhline(y=0, color='k', linestyle='-', alpha=0.3)
plt.axvline(x=0, color='k', linestyle='-', alpha=0.3)

# Add plot details
plt.title('2D Shearing of a Triangle')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.axis('equal')  # Make the plot axes equal scale

# Set plot limits to show all triangles clearly


# Show the plot
plt.show()
```

# ***2D Scaling***

```python
import matplotlib.pyplot as plt
import numpy as np

def create_triangle():
    """Create a simple triangle with vertices"""
    # Define triangle vertices (x, y coordinates)
    triangle = np.array([
        [1, 3],    # vertex 1
        [2, 5],    # vertex 2
        [3, 3]   # vertex 3
    ])
    return triangle

def scale_triangle(triangle, sx, sy):
    """Apply scaling transformation to triangle"""
    # Create scaling matrix
    scaling_matrix = np.array([
        [sx, 0],
        [0, sy]
    ])

    # Apply scaling transformation
    scaled_triangle = triangle @ scaling_matrix.T
    return scaled_triangle



def demonstrate_scaling():
    """Main function to demonstrate triangle scaling"""
    # Create original triangle
    original_triangle = create_triangle()

    # Define scaling factors (sx=1.5, sy=2.0)
    sx, sy = 1.5, 1.5

    # Apply scaling transformation
    scaled_triangle = scale_triangle(original_triangle, sx, sy)

    # Create simple side-by-side plot
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))
    fig.suptitle('2D Triangle Scaling Demonstration', fontsize=16, fontweight='bold')

    # Plot original triangle
    orig_plot = np.vstack([original_triangle, original_triangle[0]])  # Close the triangle
    ax1.plot(orig_plot[:, 0], orig_plot[:, 1], 'o-', color='blue',
            linewidth=2, markersize=8, label='Original')
    ax1.set_title('Original Triangle')
    ax1.grid(True, alpha=0.3)
    ax1.set_xlabel('X coordinate')
    ax1.set_ylabel('Y coordinate')
    ax1.legend()
    ax1.axis('equal')


    # Plot scaled triangle with original overlay
    scaled_plot = np.vstack([scaled_triangle, scaled_triangle[0]])
    ax2.plot(orig_plot[:, 0], orig_plot[:, 1], 'o-', color='blue',
            linewidth=2, markersize=6, alpha=0.5, label='Original')
    ax2.plot(scaled_plot[:, 0], scaled_plot[:, 1], 'o-', color='red',
            linewidth=2, markersize=8, label=f'Scaled (sx={sx}, sy={sy})')
    ax2.set_title(f'Scaled Triangle (sx={sx}, sy={sy})')
    ax2.grid(True, alpha=0.3)
    ax2.set_xlabel('X coordinate')
    ax2.set_ylabel('Y coordinate')
    ax2.legend()
    ax2.axis('equal')


    plt.tight_layout()
    plt.show()



if __name__ == "__main__":
    # Run the demonstration
    demonstrate_scaling()
```

# ***2D Rotation of a Line***

```python
import numpy as np
import matplotlib.pyplot as plt

def draw_line(points, color='b', label=None):
    # Draw the line by connecting two points
    plt.plot([points[0][0], points[1][0]],
             [points[0][1], points[1][1]],
             color=color,
             label=label)

def rotate_points(points, angle_degrees):
    # Convert angle to radians
    angle_rad = np.radians(angle_degrees)

    # Create rotation matrix
    rotation_matrix = np.array([[np.cos(angle_rad), -np.sin(angle_rad), 0],
                               [np.sin(angle_rad), np.cos(angle_rad), 0],
                               [0, 0, 1]])

    # Convert points to homogeneous coordinates
    homogeneous_points = np.hstack((points, np.ones((2, 1))))

    # Apply rotation
    rotated_points = np.dot(homogeneous_points, rotation_matrix.T)

    # Convert back to 2D coordinates
    return rotated_points[:, :2]

# Define the original line endpoints
original_line = np.array([
    [0, 0],    # Start point at origin
    [3, 0]     # End point on x-axis
])

# Define rotation angles
angles = [0, 45, 90, 135, 180]  # degrees

# Set up the plot
plt.figure(figsize=(10, 10))
plt.grid(True)

# Draw original and rotated lines
colors = ['b', 'r', 'g', 'm', 'c']
for angle, color in zip(angles, colors):
    rotated_line = rotate_points(original_line, angle)
    draw_line(rotated_line, color=color, label=f'Rotation {angle}°')

# Add axes lines for reference
plt.axhline(y=0, color='k', linestyle='-', alpha=0.3)
plt.axvline(x=0, color='k', linestyle='-', alpha=0.3)

# Add plot details
plt.title('2D Rotation of a Line')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.axis('equal')  # Make the plot axes equal scale

# Set plot limits to show all lines clearly


# Show the plot
plt.show()

# Optional: Animate the rotation
def animate_rotation():
    plt.figure(figsize=(10, 10))

    for angle in range(0, 361, 5):  # Rotate from 0 to 360 degrees
        plt.clf()  # Clear the current figure
        plt.grid(True)

        # Draw the rotated line
        rotated_line = rotate_points(original_line, angle)
        draw_line(rotated_line, color='r', label=f'Rotation {angle}°')

        # Add axes lines
        plt.axhline(y=0, color='k', linestyle='-', alpha=0.3)
        plt.axvline(x=0, color='k', linestyle='-', alpha=0.3)

        # Add plot details
        plt.title('2D Rotation of a Line (Animation)')
        plt.xlabel('X-axis')
        plt.ylabel('Y-axis')
        plt.legend()
        plt.axis('equal')
        plt.xlim(-4, 4)
        plt.ylim(-4, 4)

        plt.pause(0.1)  # Add a small pause to create animation effect

    plt.show()

# Uncomment the next line to see the animation
# animate_rotation()
```
